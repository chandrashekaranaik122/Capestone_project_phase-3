```python
df['Director2_Name']=df['Director2_Name'].replace("-",np.nan)
df['Director2_Name'].value_counts()
```

```python
df=df.replace('-',np.nan)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop(['Director2_Name'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder()

for i in df.columns:
    if df[i].dtypes=='object':
        df[i]=oe.fit_transform(df[i].values.reshape(-1,1))
df
```

```python
df.isnull().sum()
```

```python
df['Director1_Name'].fillna(value=round(df['Director1_Name'].mean()),inplace=True)
df['Director1_Name'].isnull().sum()
```

```python
df['Duration'].fillna(value=round(df['Duration'].mean()),inplace=True)
df['Genre_3'].fillna(value=round(df['Genre_3'].mean()),inplace=True)
df['Ratings'].fillna(value=round(df['Ratings'].mean()),inplace=True)
df['Certification'].fillna(value=round(df['Certification'].mean()),inplace=True)
df['Papularity'].fillna(value=round(df['Papularity'].mean()),inplace=True)
df['Star_1'].fillna(value=round(df['Star_1'].mean()),inplace=True)
df['Star_2'].fillna(value=round(df['Star_2'].mean()),inplace=True)
df['Star_3'].fillna(value=round(df['Star_3'].mean()),inplace=True)
df['Star_4'].fillna(value=round(df['Star_4'].mean()),inplace=True)
```
